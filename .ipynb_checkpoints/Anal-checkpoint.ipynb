```python
import numpy as np
import pandas as pandas
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.style.use('ggplot')
#mpl.style.use('classic')
import pandas as pd

quick = pd.read_csv("quick.txt", sep = '\t')
quick.head()
x_axis = [x for x in range(1000, 1000000, 1000)]


plt.plot(quick.length, quick.random, label = 'random')
plt.plot(quick.length, quick.sorted, label = 'sorted')
plt.plot(quick.length, quick.reversed, label = 'reversed')
plt.legend()
plt.xlabel('array size')
plt.ylabel('seconds')
plt.show()
```

```python
import math
def f(n):
    return n*math.log(n, 2)
def g(n):
    return math.log(n, 2)

quick_random = pd.read_csv("quick_random.txt", sep = '\t')

teste = [f(x) for x in quick_random.length]
print(teste)
#plt.plot(quick_random.length, quick_random.random, label = 'random')
#plt.plot(quick_random.length, quick_random.sorted, label = 'sorted')
#plt.plot(quick_random.length, quick_random.reversed, label = 'reversed')
plt.plot(quick_random.length,[f(x) for x in quick_random.length], label = 'nlogn')
plt.legend()
plt.ylabel('seconds')
plt.xlabel('array size')
plt.show()
```

```python
count = pd.read_csv("counting.txt", sep = '\t')
    
plt.plot(count.length, count.uniform, label = 'uniform')
plt.plot(count.length, count.notUniform, label = 'not uniform')
plt.plot(count.length, count.lowCap, label = 'low cap')
plt.plot(count.length, count.lowCap2, label = 'low cap 2')
plt.legend()
plt.ylim(0, 0.8)
plt.ylabel('seconds')
plt.xlabel('array size')
plt.show()
```

## Quicksort vs. Heapsort


```python
#Quicksort se apresenta mais eficiente para o caso aleatório
heap = pd.read_csv("heap.txt", sep = '\t')
plt.plot(quick_random.length, quick_random.random, label = 'quicksort')
plt.plot(heap.length[:-1], heap.random[:-1], label = 'heapsort')
plt.legend()
plt.ylabel('seconds')
plt.xlabel('array size')
plt.plot
plt.show()
```

```python
heap = pd.read_csv("heap.txt", sep = '\t')
plt.plot(quick.length, quick.reversed, label = 'quicksort')
plt.plot(heap.length, heap.reversed, label = 'heapsort')
plt.ylim(0, 15)
plt.legend()
plt.plot()
plt.xlabel('array size')
plt.ylabel('seconds')
plt.show()

```
