```python
import numpy as np
import pandas as pandas
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.style.use('ggplot')
#mpl.style.use('classic')
import pandas as pd

quick = pd.read_csv("quick.txt", sep = '\t')
quick.head()
x_axis = [x for x in range(1000, 1000000, 1000)]


plt.plot(quick.length, quick.random, label = 'random')
plt.plot(quick.length, quick.sorted, label = 'sorted')
plt.plot(quick.length, quick.reversed, label = 'reversed')
plt.legend()
plt.xlabel('array size')
plt.ylabel('seconds')
plt.show()
```

## Quicksort vs. Heapsort


```python
#Quicksort se apresenta mais eficiente para o caso aleatrio
heap = pd.read_csv("heap.txt", sep = '\t')
plt.plot(quick_random.length, quick_random.random, label = 'quicksort -- random pivot')
plt.plot(quick.length, quick.random, label = "quicksort -- middle pivot")
plt.plot(heap.length, heap.random, label = 'heapsort')
plt.legend()
plt.ylim(0, 20)
plt.ylabel('seconds')
plt.xlabel('array size')
plt.title("Random array")
plt.plot
plt.show()
```

```python
heap = pd.read_csv("heap.txt", sep = '\t')
plt.plot(quick.length, quick.reversed, label = 'quicksort -- middle pivot')
plt.plot(quick_random.length, quick_random.reversed, label = 'quicksort -- random pivot')
plt.plot(heap.length, heap.reversed, label = 'heapsort')
plt.ylim(0, 15)
plt.legend()
plt.plot()
plt.xlabel('array size')
plt.ylabel('seconds')
plt.title("Reversed Array")
plt.show()

```

```python
plt.plot(quick.length, quick.sorted, label = 'quicksort -- middle pivot')
plt.plot(quick_random.length, quick_random.sorted, label = 'quicksort -- random pivot')
plt.plot(heap.length, heap.sorted, label = 'heapsort')
plt.ylim(0, 15)
plt.legend()
plt.plot()
plt.xlabel('array size')
plt.ylabel('seconds')
plt.title("Sorted Array")
plt.show()
```

```python
#Quicksort se apresenta mais eficiente para o caso aleatrio
plt.plot(quick_random.length, quick_random.random, label = 'quicksort -- random pivot')
plt.plot(quick.length, quick.random, label = "quicksort -- middle pivot")
plt.plot(count.length,count.uniform, label = 'counting')
plt.legend()
#plt.ylim(0, 2)
#plt.xlim(0, 100000)
plt.ylabel('seconds')
plt.xlabel('array size')
plt.title("Random array")
plt.plot

plt.show()
```
