```python
import numpy as np
import pandas as pandas
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.style.use('ggplot')
#mpl.style.use('classic')
import pandas as pd

quick = pd.read_csv("quick.txt", sep = '\t')
quick.head()
x_axis = [x for x in range(1000, 1000000, 1000)]


plt.plot(quick.length, quick.random, linewidth = 3, label = 'random')
plt.plot(quick.length, quick.sorted, linewidth = 3, label = 'sorted')
plt.plot(quick.length, quick.reversed, linewidth= 3, label = 'reversed')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('array size')
plt.ylabel('seconds')
plt.title("Quicksort - Middle Pivot")
plt.savefig("quicksort.eps", bbox_inches = 'tight')
plt.show()
```

```python
import math
def f(n):
    return n*math.log(n, 2)

quick_random = pd.read_csv("quick_random.txt", sep = '\t')

plt.plot(quick_random.length, quick_random.random, linewidth = 3, label = 'random')
plt.plot(quick_random.length, quick_random.sorted, linewidth = 3, label = 'sorted')
plt.plot(quick_random.length, quick_random.reversed, linewidth = 3, label = 'reversed')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.ylabel('seconds')
plt.xlabel('array size')
plt.title("Quicksort - Random Pivot")
plt.savefig("quicksort-random.eps", bbox_inches = 'tight')
plt.show()
```

```python
count = pd.read_csv("counting.txt", sep = '\t')
    
plt.plot(count.length, count.uniform, linewidth = 3,label = 'uniform')
plt.plot(count.length, count.random, linewidth = 3, label = 'random')
plt.plot(count.length, count.lowCap, linewidth = 3, label = 'low cap')
plt.plot(count.length, count.unbalanced, linewidth = 3, label = 'unbalanced')
plt.legend()
plt.ylim(0, 0.5)
plt.ylabel('seconds')
plt.xlabel('array size')
plt.title('counting')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

## Quicksort vs. Heapsort


```python
#Quicksort se apresenta mais eficiente para o caso aleatrio
heap = pd.read_csv("heap.txt", sep = '\t')
plt.plot(quick_random.length, quick_random.random, linewidth = 3, label = 'quicksort -- random pivot')
plt.plot(quick.length, quick.random, linewidth = 3, label = "quicksort -- middle pivot")
plt.plot(heap.length, heap.random, linewidth = 3, label = 'heapsort')
plt.legend()
plt.ylim(0, 20)
plt.ylabel('seconds')
plt.xlabel('array size')
plt.title("Random array")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.plot
plt.show()
```

```python
heap = pd.read_csv("heap.txt", sep = '\t')
plt.plot(quick.length, quick.reversed, linewidth = 3,label = 'quicksort -- middle pivot')
plt.plot(quick_random.length, quick_random.reversed, linewidth = 3, label = 'quicksort -- random pivot')
plt.plot(heap.length, heap.reversed, linewidth = 3, label = 'heapsort')
plt.ylim(0, 15)
plt.legend()
plt.plot()
plt.xlabel('array size')
plt.ylabel('seconds')
plt.title("Reversed Array")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()

```

```python
plt.plot(quick.length, quick.sorted, linewidth = 3,label = 'quicksort -- middle pivot')
plt.plot(quick_random.length, quick_random.sorted, linewidth = 3, label = 'quicksort -- random pivot')
plt.plot(heap.length, heap.sorted, linewidth = 3, label = 'heapsort')
plt.ylim(0, 15)
plt.legend()
plt.plot()
plt.xlabel('array size')
plt.ylabel('seconds')
plt.title("Sorted Array")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
